```python
import os

path_dir = 'data0_m4a/'
file_list = os.listdir(path_dir)

print(file_list)
```

```python
no_space = [file.replace(" ", "")[:file.index(".")-1] for file in file_list]

print(no_space[10])
```

```python
output_dir = "data0_wav/"

for file in file_list:
    os.system('ffmpeg -i  "%s%s"  "%s%s.wav"' % (path_dir, file, output_dir, file.replace(" ", "")[:file.index(".")-1]))
```
